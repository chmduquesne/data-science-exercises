```python
from vega_datasets import data
from quickviz import visualize as v
import pandas as pd
import numpy as np
df = data.cars()
```

```python
df = df.dropna(subset=['Miles_per_Gallon', 'Horsepower'])

# return the middles of the bins
def bin_mids(a, bins=10):
    return list(map(lambda i: i.mid, pd.cut(a, bins=bins)))

df["Miles_per_Gallon (binned)"] = bin_mids(df.Miles_per_Gallon)
df["Horsepower (binned)"] = bin_mids(df.Horsepower)
```

```python
groups = df[["Horsepower (binned)", "Miles_per_Gallon (binned)", "Name"]] \
            .groupby(["Horsepower (binned)", "Miles_per_Gallon (binned)"], as_index=False) \
            .agg("count")
```

```python
plot = v(groups)
plot.load({'kwargs': {'size': 'Name',
  'x': 'Horsepower (binned)',
  'y': 'Miles_per_Gallon (binned)'},
 'plot': 'relplot'})
```
