```python
from vega_datasets import data
from quickviz import visualize as v
import pandas as pd
import numpy as np
df = data.cars()
```

```python
# return the middles of the bins
def bins_centers(a, bins=10):
    return list(map(lambda i: i.mid if hasattr(i, 'mid') else i, pd.cut(a, bins=bins)))

df["Miles_per_Gallon (binned)"] = bins_centers(df.Miles_per_Gallon)
df["Horsepower (binned)"] = bins_centers(df.Horsepower)
```

```python
groups = df.groupby(["Horsepower (binned)", "Miles_per_Gallon (binned)"]).agg('size').reset_index(name="Count")
```

```python
plot = v(groups)
plot.load({'kwargs': {'size': 'Count',
  'x': 'Horsepower (binned)',
  'y': 'Miles_per_Gallon (binned)'},
 'plot': 'relplot'})
```
